```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# Define the base directory and file paths
base_dir = Path.cwd().parent.parent
json_file = base_dir / "data" / "jsonpqt" / "ajac0010_metadata.json"
parquet_file = base_dir / "data" / "jsonpqt" / "ajac0010_observations.parquet"

# Read JSON metadata
with open(json_file, 'r') as f:
    metadata = json.load(f)

# Load observation data
df = pd.read_parquet(parquet_file)

# Select a specific satellite and channel
satellite = "G02"
channel = "1"

# Filter data for the selected satellite and channel
filtered_data = df[(df["sat"] == satellite) & (df["channel"] == channel)]


# Plot SNR data
plt.figure(figsize=(10, 6))
plt.plot(filtered_data["mjd"], filtered_data["snr"], label=f"{satellite} L1")
plt.xlabel("Time")
plt.ylabel("SNR")
plt.title(f"SNR for {satellite} L1")
plt.legend()
plt.show()
```
